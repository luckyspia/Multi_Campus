## Loss Function

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
y = np.array([0, 1, 0, 0])

yhat_1 = np.array([0.1, 0.7, 0.2, 0])
yhat_2 = np.array([0.1, 0.2, 0.7, 0])

N = y.size
```

#### 평균 제곱 오차 (MSE: Mean Squared Error)

```python
mse_1 = np.sum(np.square(yhat_1 - y)) / N
mse_2 = np.sum(np.square(yhat_2 - y)) / N
mse_1, mse_2
```

#### 평균 제곱근 오차 (RMSE: Root Mean Squared Error)

```python
rmse_1 = np.sqrt(mse_1)
rmse_2 = np.sqrt(mse_2)
rmse_1, rmse_2
```

#### 평균 절대 오차 (MAE: Mean Absolute Error)

```python
mae_1 = np.sum(np.abs(yhat_1 - y)) / N
mae_2 = np.sum(np.abs(yhat_2 - y)) / N
mae_1, mae_2
```

#### 범주형 교차 엔트로피 (Categorical Cross Entropy)

```python
def cross_entropy_error(yhat, ytrue):
    if yhat.ndim == 1:
        yhat = yhat.reshape(1, -1)
        ytrue = ytrue.reshape(1, -1)
    if yhat.size == ytrue.size:
        ytrue = ytrue.argmax(axis=1)
    batch_size = yhat.shape[0]
    return -np.sum(np.log(yhat[np.arange(batch_size), ytrue])) / batch_size
```

```python
print('step 1:',yhat_1.ndim)
print('step 2:',yhat_1.reshape(1, -1))
print('step 3:',y.reshape(1,-1).argmax(axis=1))
print('step 4:',yhat_1.shape[0] )
```

```python
cce1 = cross_entropy_error(yhat_1, y)
cce2 = cross_entropy_error(yhat_2, y)
cce1, cce2
```

#### 이진형 교차 엔트로피 (Binary Cross Entropy)

```python
# log probability Func
xs = np.linspace(0.001, 0.999, 200)
plt.figure(figsize=(12,8))
plt.subplot(221)
plt.plot(xs, np.log(xs)), plt.title('log x'), plt.grid()
plt.subplot(222)
plt.plot(xs, -np.log(xs)), plt.title('-log x'), plt.grid()
plt.subplot(223)
plt.plot(xs, np.log(1 - xs)), plt.title('log (1 - x)'), plt.grid()
plt.subplot(224)
plt.plot(xs, -np.log(1 - xs)), plt.title('-log (1 -x)'), plt.grid()
plt.show()
```

```python
# Binary CrossEntropy Graph
# -ylogx - (1-y)log(1-x)
plt.plot(xs, -np.log(xs), lw=2, label='-ylogx')
plt.plot(xs, -np.log(1-xs), lw=2, label='-(1-y)log(1-x)')
plt.title('-ylogx - (1-y)log(1-x)')
plt.grid(), plt.legend()
plt.show()
```
