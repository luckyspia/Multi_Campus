```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()
df = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
df['target'] = diabetes.target
df.head()
```

```python
# N(0, 0.0476^2) 으로 이미 스케일링된 값
df.describe()
```

```python
print(diabetes.DESCR)
```

```python
import seaborn as sns
_, axs = plt.subplots(2, 5, figsize=(18, 8))
for idx, feature in enumerate(df.columns[:-1]):
    row, col = idx // 5, idx % 5
    sns.regplot(x=feature, y='target', data=df, ax=axs[row][col])
```

#### BMI vs. target, BP vs. target... (Feature vs. target)

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
for feature in df.columns[:-1]:
    X = df[feature].values.reshape(-1, 1)
    lr.fit(X, diabetes.target)
    score = lr.score(X, diabetes.target)
    print(f'{feature}: {score:.4f}')
```

```python
from sklearn.model_selection import train_test_split
# Regression 은 Stratify를 적용할 수 없음
X_train, X_test, y_train, y_test = train_test_split(
    df.bmi.values.reshape(-1, 1), diabetes.target, test_size=0.1, random_state=2022
)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
# 계수, 절편, 잔차제곱합
lr.coef_, lr.intercept_, lr._residues
```

```python
from sklearn.metrics import r2_score
pred = lr.predict(X_test)
r2_score(y_test, pred)
```

```python
# Mean Squared Error = 절대적인 값이 아니므로, 모델의 성능을 판단할 수 없지만 특성 간 비교하기 좋음
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, pred)
rmse = np.sqrt(mse)
mse, rmse
```

```python
xs = np.array([-0.1, 0.18])
ys = xs * lr.coef_ + lr.intercept_
```

```python
plt.figure(figsize=(12, 8))
plt.scatter(X_train, y_train)
plt.plot(xs, ys, 'r-', lw=3)
plt.grid()
plt.xlabel("BMI")
plt.ylabel("target")
plt.title('BMI vs target', fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
plt.scatter(X_train, y_train, label='Train')
plt.scatter(X_test, y_test, marker='^', label='Test')
plt.plot(xs, ys, 'r-', lw=3)
plt.grid()
plt.legend()
plt.xlabel("BMI")
plt.ylabel("target")
plt.title('BMI vs target', fontsize=15)
plt.show()
```
